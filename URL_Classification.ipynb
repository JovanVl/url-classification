```python
def get_X(data, l, r):
    matrices = [data[f'Day{i}']['data'][0][0] for i in range(l, r)]
    X = vstack(matrices, format='csc')
    
    return X

def get_Y(data, l, r):
    y = []
    for i in range(l, r):
        day_i_labels = data[f'Day{i}']['labels'][0][0]
        for j in range(0, len(day_i_labels)):
            y += day_i_labels[j].tolist()

    return y
```

```python
class NaiveBayes:
    def __init__(self, csc_matrix, labels, real_features):
        self.X = csc_matrix[0][0]
        self.num_of_urls = csc_matrix[0][0].shape[0]
        self.num_of_features = csc_matrix[0][0].shape[1]
        self.Y = self.get_Y(labels)
        self.real_features = np.array([i-1 for i in self.get_real_features(real_features)])
        self.gnb_clf = GaussianNB()
        self.bnb_clf = BernoulliNB()
        self.X_train, self.X_test, self.y_train, self.y_test = self.split_train_test()
        
    def get_Y(self, labels):
        y = []
        for i in range(0, self.num_of_urls):
            y += labels[0][0][i].tolist()

        return y

    def get_real_features(self, real_features):
        real_features = []
        for elem in feature_types:
            real_features += elem.tolist()

        return real_features
    
    def split_train_test(self):
        X_train, X_test, y_train, y_test = train_test_split(self.X, self.Y, test_size=0.2, random_state=42)
        return X_train, X_test, y_train, y_test

    def X_real_values(self, X):
        X_real = np.empty((0, len(real_features)), float)
        for i in range(0, X.shape[0]):
            row = X.getrow(i).toarray()[0]
            real_row = np.empty((0, 1), float)
            for val in self.real_features:
               real_row = np.append(real_row, row[val - 1])

            X_real = np.vstack((X_real, real_row))

        return X_real

    def X_binary_values(self, X):
        columns = np.arange(self.num_of_features)
        binary_features_indices = np.setdiff1d(columns, self.real_features)
        matrix_with_binary_features = X[:, binary_features_indices]
        
        return matrix_with_binary_features

    def fit(self):
        self.gnb_clf.fit(self.X_real_values(self.X_train), self.y_train)
        self.bnb_clf.fit(self.X_binary_values(self.X_train), self.y_train)

    def predict(self, X_test):
        log_prob_gauss = self.gnb_clf.predict_log_proba(self.X_real_values(X_test))
        log_prob_bernoulli = self.bnb_clf.predict_log_proba(self.X_binary_values(X_test))
        log_prob = log_prob_gauss + log_prob_bernoulli
        
        return np.argmax(log_prob, axis=1)
```

```python
class SVM:
    def __init__(self, csc_matrix, labels, real_features):
        self.X = csc_matrix[0][0]
        self.num_of_urls = csc_matrix[0][0].shape[0]
        self.num_of_features = csc_matrix[0][0].shape[1]
        self.Y = self.get_Y(labels)
        self.real_features = np.array([i-1 for i in self.get_real_features(real_features)])
        self.X_train, self.X_test, self.y_train, self.y_test = self.split_train_test()
        
    def get_Y(self, labels):
        y = []
        for i in range(0, self.num_of_urls):
            y += labels[0][0][i].tolist()

        return y

    def get_real_features(self, real_features):
        real_features = []
        for elem in feature_types:
            real_features += elem.tolist()

        return real_features
    
    def split_train_test(self):
        X_train, X_test, y_train, y_test = train_test_split(self.X, self.Y, test_size=0.3, random_state=42)
        return X_train, X_test, y_train, y_test

    def fit(self):
        model = SVC(kernel='linear')
        model.fit(self.X_train, self.y_train)
        
        return model
```

```python
class LR:
    def __init__(self, csc_matrix, labels, real_features):
        self.X = csc_matrix[0][0]
        self.num_of_urls = csc_matrix[0][0].shape[0]
        self.num_of_features = csc_matrix[0][0].shape[1]
        self.Y = self.get_Y(labels)
        self.real_features = np.array([i-1 for i in self.get_real_features(real_features)])
        self.X_train, self.X_test, self.y_train, self.y_test = self.split_train_test()
        
    def get_Y(self, labels):
        y = []
        for i in range(0, self.num_of_urls):
            y += labels[0][0][i].tolist()

        return y

    def get_real_features(self, real_features):
        real_features = []
        for elem in feature_types:
            real_features += elem.tolist()

        return real_features
    
    def split_train_test(self):
        X_train, X_test, y_train, y_test = train_test_split(self.X, self.Y, test_size=0.1, random_state=42)
        return X_train, X_test, y_train, y_test

    def fit(self):
        model = LogisticRegression(max_iter=1000, solver='lbfgs', random_state=42)
        model.fit(self.X_train, self.y_train)
        
        return model
```
